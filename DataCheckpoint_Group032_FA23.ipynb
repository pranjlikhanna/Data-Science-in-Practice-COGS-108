# Research Question

How does the neutrality and subjectivity of job postings for “Data Analyst” or “Data Scientist” positions on Glassdoor correlate with the company’s overall Glassdoor rating (out of 5 stars) for medium to large-sized US companies with 500 to 5000 employees?

# Background and Prior Work

Job postings provide key information about an open job position to serve the purpose of recruiting new employees. From a job seeker’s perspective, the job posting creates the first impression of the employer and serves a crucial role in judging potential company fit. Based on content, structure, tone, and length, job seekers can form assumptions about a company’s work culture, values, management, and other aspects. 

Given that job seekers make speculations about a company from the job posting, our team was interested in discovering whether these speculations have any statistical basis. As undergraduate students who read many job postings in our search for work, we were most interested in the element of tone. According to LinkedIn Talent Solutions, applicants are 2-4x less likely to apply for a job posting with a tone that doesn't match the perceived company culture [1]. If the tone of a job posting is not genuinely reflective of satisfaction within a company, then job seekers should adjust their perception to account for these insights. 

Using existing data, we aim to determine if there is a statistical relationship between the degree of neutrality and subjectivity expressed in job postings and the overall satisfaction of employees working at the company. This research focuses on “Data Analyst/Scientist” positions, as they closely align with the interests of the research team. To collect data on job postings, we use existing raw datasets scraped from Glassdoor found on Kaggle. For the neutrality and subjectivity analysis, we use Vader to determine the degree of neutrality and TextBlob to analyze subjectivity [2]. 

Prior to data collection and analysis, we conducted a literature review of past studies to gain a better understanding of research on the relationship between job postings, company performance, and public perception. 

A study conducted by Stanford University [3], converted qualitative reviews into quantitative metrics and used them to predict the corresponding ratings on Glassdoor. The research findings indicated that their model had a 46.4% accuracy in using the sentiment of a review to predict the ratings on a scale from 1 to 5 stars. While our study will be conducted through a different method, the conclusion from this existing study serves as a reference for our study outcome, and further comparison and analysis could be done based on both conclusions. Another study looked into how the tone of job postings affects applicants’ impression of the employers and likelihood to apply [4]. The study concluded that job postings with a generic tone are more likely to attract applicants than those with a formal or casual tone, which implies that the tone of the job posting is directly related to applicants’ impressions of the position and employer.

Based on our existing knowledge and experience in analyzing job postings and the available literature, we hypothesize that within our proposed research scope, there will be a correlation between the degree of neutrality and the degree of subjectivity of the job postings, and the overall company rating.

# Hypothesis


### Neutrality

**Null Hypothesis (H0):** There is no statistically significant correlation between the neutrality score assigned by VADER to job postings for "Data Analyst" or "Data Scientist" positions and the corresponding company's ratings on Glassdoor.

**Alternative Hypothesis (H1):** There is a statistically significant positive correlation between the neutrality score assigned by VADER to job postings for "Data Analyst" or "Data Scientist" positions and the corresponding company's ratings on Glassdoor.

This implies that as the neutrality score of job postings increases, there is an expected positive impact on the company's overall rating on Glassdoor. We believe that companies that present their job-opportunities in a balanced tone are more likely to be perceived positively on Glassdoor.


### Subjectivity
**Null Hypothesis (H0):** There is no statistically significant correlation between the subjectivity score assigned by TextBlob to job postings for “Data Analyst” or “Data Scientist” positions and the corresponding company’s ratings on Glassdoor.

**Alternative Hypothesis (H1):** There is a statistically significant negative correlation between the subjectivity score assigned by TextBlob to job postings for “Data Analyst” or “Data Scientist” positions and the corresponding company’s ratings on Glassdoor.

We believe that a non-subjective and more objective tone is usually associated with higher professionalism and higher credibility. Thus, we believe that job postings that have lower subjectivity scores would have higher company ratings on Glassdoor.

## Data Cleaning

```python
# perform cleaning on all three datasets
df_list = [df1, df2, df3]

for count, df in enumerate(df_list):
    # Change the name of columns
    df = df[['Job Title', 'Job Description', 'Rating', 'Company Name', 'Location', 'Size']]
    df = df.rename({'Job Title':'Job Type', 
                    'Job Description':'Job Posting', 
                    'Company Name':'Company'},
                    axis='columns')
    # Remove the rows that have missing data
    for columns in df.columns:
        df = df[df[columns]!=-1]

    # Make sure only containing data scientist/analyst data
    df['Job Type'] = df['Job Type'].str.lower()
    df = df[df['Job Type'].str.contains('data scientist|data analyst')]
    df.iloc[:,0] = 'Data Scientist/Analyst'

    # Formatting the columns
    df['Company'] = df['Company'].str.replace(r'\n', '').str.replace('\d+', '').str.replace('.', '');
    df['Job Posting'] = df['Job Posting'].str.replace(r'\n', ' ')

    # Check whether the company is in the USA
    df = df.reset_index(drop=True)
    drop_list = []
    states_in_united_states = [ 'AK', 'AL', 'AR', 'AZ', 'CA', 'CO', 'CT', 'DC', 'DE', 'FL', 'GA',
                                'HI', 'IA', 'ID', 'IL', 'IN', 'KS', 'KY', 'LA', 'MA', 'MD', 'ME',
                                'MI', 'MN', 'MO', 'MS', 'MT', 'NC', 'ND', 'NE', 'NH', 'NJ', 'NM',
                                'NV', 'NY', 'OH', 'OK', 'OR', 'PA', 'RI', 'SC', 'SD', 'TN', 'TX',
                                'UT', 'VA', 'VT', 'WA', 'WI', 'WV', 'WY']
    # check whether the location of company is in the USA or not, update location if yes otherwise drop it
    for index, location in enumerate(df['Location']):
        flag = False
        location_list = location.replace(' ', '').split(',')
        if len(location_list) == 2:
            for state in states_in_united_states:
                if location_list[1] == state:
                    flag = True
                    df['Location'][index] = state
                    break
        else: 
            drop_list.append(index)
            continue

        if flag == False:
            drop_list.append(index)
    # drop rows that contain company that is not in the USA
    df = df.drop(df.index[drop_list])

    # Find company within a criteria and categorize the company by its size
    drop_list = []
    df = df.reset_index(drop=True)
    for index, size in enumerate(df['Size']):
        digit_list = [i for i in size.split() if i.isdigit()]

        if len(digit_list) == 2:
            if digit_list[0] == '501' and digit_list[1] == '1000':
                df['Size'][index] = 'Middle'
            elif digit_list[0] == '1001' and digit_list[1] == '5000':
                df['Size'][index] = 'Relatively Large'
            else:
                drop_list.append(index)
        else:
            drop_list.append(index)
    #drop rows that does not satisfy the size criteria
    df = df.drop(df.index[drop_list])
    
    # update each dataframe
    df_list[count] = df
```

```python
# merge all datasets and remove rows that have duplicated company
result_dataset = pd.concat(df_list)
result_dataset = result_dataset.drop_duplicates(subset=['Company'])
```

```python
# display few first rows after cleaning the dataset
result_dataset = result_dataset.reset_index(drop=True)
result_dataset.head()
```

## Sentiment Analysis

```python
# get the sentiment analysis and create columns for them
analyser = SentimentIntensityAnalyzer()
result_dataset[['Neutrality', 'Subjectivity']] = [0.0, 0.0]

# iterate through the entire dataset to get percent for each row
for index, post in enumerate(result_dataset['Job Posting']):
    resulted_sent = analyser.polarity_scores(post)
    result_dataset['Neutrality'][index] = resulted_sent['neu']
    blob = TextBlob(post)
    result_dataset['Subjectivity'][index] = round(blob.sentiment.subjectivity, 3)

result_dataset.head()
```
