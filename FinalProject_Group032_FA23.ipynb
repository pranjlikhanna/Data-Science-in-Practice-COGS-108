# Overview

A job posting is crucial for a job seeker in deciding their suitability for a role, as it provides insights into the job requirements, company culture, expected responsibilities, benefits, and more. In addition to the content of the job posting, the tone can have an impact on a job seeker’s perception of the job position and the hiring company. As a team of prospective job seekers, we were interested in understanding whether the tone of a job posting correlates with a company’s rating to better inform our judgments during the job hunting process. 

Using sentiment analysis tools, we examined how neutrality and subjectivity in Data Scientist job postings correlate with ratings of medium to large-sized American companies, ranging from 500 to 5000 employees. We hypothesized that job postings reflecting a more neutral and less subjective tone would correlate with higher company ratings. Although our initial findings suggested a weak correlation, further nuanced analysis is needed to fully understand the dynamics between job posting tone and company rating.

# Research Question

How does the neutrality and subjectivity of job postings for “Data Analyst” or “Data Scientist” positions on Glassdoor correlate with the company’s overall Glassdoor rating (out of 5 stars) for medium to large-sized US companies with 500 to 5000 employees?

# Hypothesis


### Neutrality

**Null Hypothesis (H0):** There is no statistically significant correlation between the neutrality score assigned by VADER to job postings for "Data Analyst" or "Data Scientist" positions and the corresponding company's ratings on Glassdoor.

**Alternative Hypothesis (H1):** There is a statistically significant positive correlation between the neutrality score assigned by VADER to job postings for "Data Analyst" or "Data Scientist" positions and the corresponding company's ratings on Glassdoor.

This implies that as the neutrality score of job postings increases, there is an expected positive impact on the company's overall rating on Glassdoor. We believe that companies that present their job-opportunities in a balanced tone are more likely to be perceived positively on Glassdoor.


### Subjectivity
**Null Hypothesis (H0):** There is no statistically significant correlation between the subjectivity score assigned by TextBlob to job postings for “Data Analyst” or “Data Scientist” positions and the corresponding company’s ratings on Glassdoor.

**Alternative Hypothesis (H1):** There is a statistically significant negative correlation between the subjectivity score assigned by TextBlob to job postings for “Data Analyst” or “Data Scientist” positions and the corresponding company’s ratings on Glassdoor.

We believe that a non-subjective and more objective tone is usually associated with higher professionalism and higher credibility. Thus, we believe that job postings that have lower subjectivity scores would have higher company ratings on Glassdoor.

## Data Cleaning

```python
# perform cleaning on all three datasets
df_list = [df1, df2, df3]

for count, df in enumerate(df_list):
    # Change the name of columns
    df = df[['Job Title', 'Salary Estimate', 'Job Description', 'Rating', 'Company Name', 'Location', 'Size']]
    df = df.rename({'Job Title':'Job Type', 
                    'Job Description':'Job Posting', 
                    'Company Name':'Company'},
                    axis='columns')
    # Remove the rows that have missing data
    for columns in df.columns:
        df = df[df[columns]!=-1]

    # Make sure only containing data scientist/analyst data
    df['Job Type'] = df['Job Type'].str.lower()
    df = df[df['Job Type'].str.contains('data scientist|data analyst')]
    df.iloc[:,0] = 'Data Scientist/Analyst'

    # Formatting the columns
    df['Company'] = df['Company'].str.replace(r'\n', '').str.replace('\d+', '').str.replace('.', '');
    df['Job Posting'] = df['Job Posting'].str.replace(r'\n', ' ')

    # Check whether the company is in the USA
    df = df.reset_index(drop=True)
    drop_list = []
    states_in_united_states = [ 'AK', 'AL', 'AR', 'AZ', 'CA', 'CO', 'CT', 'DC', 'DE', 'FL', 'GA',
                                'HI', 'IA', 'ID', 'IL', 'IN', 'KS', 'KY', 'LA', 'MA', 'MD', 'ME',
                                'MI', 'MN', 'MO', 'MS', 'MT', 'NC', 'ND', 'NE', 'NH', 'NJ', 'NM',
                                'NV', 'NY', 'OH', 'OK', 'OR', 'PA', 'RI', 'SC', 'SD', 'TN', 'TX',
                                'UT', 'VA', 'VT', 'WA', 'WI', 'WV', 'WY']
    # check whether the location of company is in the USA or not, update location if yes otherwise drop it
    for index, location in enumerate(df['Location']):
        flag = False
        location_list = location.replace(' ', '').split(',')
        if len(location_list) == 2:
            for state in states_in_united_states:
                if location_list[1] == state:
                    flag = True
                    df['Location'][index] = state
                    break
        else: 
            drop_list.append(index)
            continue

        if flag == False:
            drop_list.append(index)
    # drop rows that contain company that is not in the USA
    df = df.drop(df.index[drop_list])

    # Find company within a criteria and categorize the company by its size
    drop_list = []
    df = df.reset_index(drop=True)
    for index, size in enumerate(df['Size']):
        digit_list = [i for i in size.split() if i.isdigit()]

        if len(digit_list) == 2:
            if digit_list[0] == '501' and digit_list[1] == '1000':
                df['Size'][index] = 'Medium'
            elif digit_list[0] == '1001' and digit_list[1] == '5000':
                df['Size'][index] = 'Large'
            else:
                drop_list.append(index)
        else:
            drop_list.append(index)
    #drop rows that does not satisfy the size criteria
    df = df.drop(df.index[drop_list])
    
    # update each dataframe
    df_list[count] = df
```

```python
# merge all datasets and remove rows that have duplicated company
result_dataset = pd.concat(df_list)
result_dataset = result_dataset.drop_duplicates(subset=['Company'])
```

## Sentiment Analysis

```python
# get the sentiment analysis and create columns for them
analyser = SentimentIntensityAnalyzer()
result_dataset[['Neutrality', 'Subjectivity']] = [0.0, 0.0]

# iterate through the entire dataset to get percent for each row
for index, post in enumerate(result_dataset['Job Posting']):
    resulted_sent = analyser.polarity_scores(post)
    result_dataset['Neutrality'][index] = resulted_sent['neu']
    blob = TextBlob(post)
    result_dataset['Subjectivity'][index] = round(blob.sentiment.subjectivity, 3)

result_dataset.head()
```

Sadly, these variables have no clear correlation even after being split by the company size. However, we have seen a possible increasing average rating with an increase in salary estimate in the EDA; hence, we are going to explore this relationship.

```python
# check the relationship between salary estimate and rating using ANOVA oneway test
f_oneway(result_dataset[result_dataset['Salary Estimate'] == '0K-50K'].Rating,
         result_dataset[result_dataset['Salary Estimate'] == '51K-100K'].Rating,
         result_dataset[result_dataset['Salary Estimate'] == '101K-150K'].Rating,
         result_dataset[result_dataset['Salary Estimate'] == '151K-200K'].Rating,
         result_dataset[result_dataset['Salary Estimate'] == 'more than 201K'].Rating)
```

It seems like there is indeed a relationship between salary estimate and rating as the p_value is really small. And let's look into more details using the linear regression model.

```python
# perform linear regression model between salary esitmate and rating
outcome, predictors = patsy.dmatrices('Rating ~ Q("Salary Estimate") - 1', result_dataset)
model = sm.OLS(outcome, predictors)
results = model.fit()
print(results.summary())
```

Indeed, even though there is a slight decrease in the coefficient in 51K-100K, there is a general increase in the coefficient with an increase in salary estimate, which means that there is a positive correlation between rating and salary estimate. However, since we previously used the ANOVA one-way test, we are assuming each salary estimate level has a normal population distribution and these distributions have the same variance. This is not actually quite true in our dataset as demonstrated above in the boxplot in EDA; hence, we may have a greater chance of falsely rejecting the true null hypothesis.

We used pairplots to find whether company size affected rating, from which we concluded that large companies would have a greater average in rating, neutrality, and subjectivity than medium-sized companies. However, no strong correlation was found between rating and neutrality, nor between rating and subjectivity from the scatterplots, thus we decided to conduct further parametric tests. We also found a trend that the average rating would tend to be higher with increasing salaries, which was further supported by the linear regression model.

Moving on to inferential analysis, we conducted pearsonr tests to find numerical support for our findings in the EDA. We found a slightly negative linear correlation between rating and neutrality with a p-value of 0.692, thus we concluded that the correlation is not statistically significant. A slightly positive linear correlation exists between rating and subjectivity, but the p-value of 0.374 concluded that the correlation is not statistically significant. Therefore, we failed to reject the null hypothesis for both neutrality and subjectivity.

Given the conclusion, we attempted to find other correlations that potentially affect company ratings. Continuing with our findings in the EDA, we looked into the correlation between rating and company size. By conducting pearsonr tests for both large and medium-sized companies on both neutrality and subjectivity, we found no clear correlation for any combination as the p-values of 4 tests were all larger than 0.05. Changing our focus once again, we looked into potential relationships between rating and salary estimate. Using the ANOVA one-way test and linear regression model, we found a clear positive correlation between the two variables with a p-value of 0.0007, noticing a slight decrease in coefficient at the 51K-100K salary estimate level. It is necessary to note that both tests were conducted based on the assumption of having a normal population distribution and the same variance across all salary estimate levels, while in reality, the dataset didn’t show such traits. Therefore, we might have falsely rejected the true null hypothesis.	

However, our study is not without its limitations. The reliance on sentiment analysis algorithms introduces potential biases, and the exclusive focus on Data Analyst/Scientist roles in the United States limits the generalizability of our findings. Looking ahead, future improvements should aim to mitigate biases in sentiment analysis, expand the scope of roles and industries analyzed, and explore company cultures in diverse global contexts.
